## Logistic Regression 

Concept: 
1. A binary classification model, input the result of linear regression 
2. Output Range(0, 1)

Core Idea:
1. Using linear model $f(x) = w^{T}x + b$, get a value from features.
2. Convert the value into a probability value via sigmoid function $\frac{1}{1 + e^{-x}}$
    - if p > 0.5, then the output would be classfication 1
    - if p < 0.5, then the output would be classification 0


Bernoulli/Binomial Probability: 
$$
P(y_i \mid x_i) = \hat{y}_i^{\,y_i} (1 - \hat{y}_i)^{\,1-y_i}
$$

MLE:
$$
L(w, b) 
= \prod_{i=1}^{m} 
\hat{y}_i^{\,y_i} (1 - \hat{y}_i)^{\,1-y_i}
$$

Because,
$$
log(a b c) = log(a) + log(b) + log(c)
$$



Loss function in Logistic Regression:
$$
L = -\frac{1}{m}\sum_{i=1}^{m} \left[ y^{(i)}\log(\hat{y}^{(i)}) + (1 - y^{(i)})\log(1-\hat{y}^{(i)}) \right]
$$


```python
import kagglehub
path = kagglehub.dataset_download("mariolisboa/breast-cancer-wisconsin-original-data-set")

import os
import pandas as pd
print(os.listdir(path))

file_path = os.path.join(path, 'breast_cancer_bd.csv')
df = pd.read_csv(file_path)
```

```python
print(df.columns)
print(df.head())
```

```python
df.info()
df.shape
```

### Clean Data

```python
import numpy as np
df = df.replace('?', np.nan)
df.dropna(axis = 0, how = 'any', inplace = True)
df.shape
```

### Import Packages

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
# Split data and target
print(df.shape)
data = df.iloc[:, 1:10]
target = df.iloc[:, 10:11]
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size = 0.2, random_state = 42)
```

### Feature Engineering


```python
transfer = StandardScaler()
x_train = transfer.fit_transform(x_train)
x_test = transfer.transform(x_test)
```

## Model Tranining

```python
estimator = LogisticRegression()
estimator.fit(x_train, y_train['Class'])
# note that fit(x: must be 2D(DataFrame), y: must be 1D(Series))
y_pre = estimator.predict(x_test)
```

```python
print(accuracy_score(y_test, y_pre))
```

## Confusion Matrix

|                           | **Predicted Positive** | **Predicted Negative** |
| ------------------------- | ------------------------------ | ------------------------------ |
| **Actual Positive** | True Positive (**TP**)         | False Negative (**FN**)        |
| **Actual Negative** | False Positive (**FP**)        | True Negative (**TN**)         |

For example: 
sample size: 10, 6 positive, 4 negative, we assume positive as true

Model A: 3 positive, 4 negative correct 

Actual Positive: 6, True Positive (TP): 3, False Negative: 3

Actual Negative: 4, False Positive (FP): 0, True Negative 4

*Precesion:*
$$
\frac{TP}{TP + FP}
$$

*Recall:*
$$
\frac{TP}{TP + FN}
$$

*F1 Score:*
$$
2\times \frac{Precision \times Recall}{Precision + Recall}
$$

Precision: evaluate the accuracy of the model

Recall: evaluate the ability to check all relevant instances

F1: evaluate classifier's accuracy combining Precision and Recall

```python
import pandas as pd
from sklearn.metrics import confusion_matrix, precision_score, recall_score
# For example: 
# sample size: 10, 6 positive, 4 negative, we assume positive as true

# Model A: 3 positive, 4 negative correct 

y_train = ['+', '+', '+', '+', '+', '+', '-', '-', '-', '-']
y_predA = ['+', '+', '+', '-', '-', '-', '-', '-', '-', '-']
label = ['+', '-']
cm = confusion_matrix(y_train, y_predA, labels = label)
df = pd.DataFrame(cm, index = label, columns = label)
print(df)
print(f'Precision score is {precision_score(y_train, y_predA, pos_label='+')}')
print(f'Recall score is {recall_score(y_train, y_predA, pos_label='+')}')
```
